# Exploratory Data Analysis

### Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

# Visualization libraries
from matplotlib import pyplot as plt     # same as 'import matplotlib.pyplot as plt'
# Plot the graph embedded in the notebook
%matplotlib inline
import seaborn as sns
```

### Load data

```python
# Load Boston housing data set
boston = load_boston()
```

```python
print(boston.DESCR)
```

### Setup dataframe

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
```

### Visualize data distributions

```python
# Plot the distribution of the target variable
sns.set(rc={'figure.figsize':(12,6)})
plt.hist(df['PRICE'], bins=50)
plt.xlabel("House prices in $1000")
plt.show()
```

```python
avg_price = df['PRICE'].mean()
avg_price
```

```python
# Plot variable CRIM against PRICE
plt.figure(figsize=(12, 6))
# plt.plot(df[['CRIM']], df['PRICE'], 'bo')     # blue circles     # Does not work with latest Pandas
plt.plot(df.iloc[:,0], df.iloc[:,13], 'bo')     # blue circles
plt.title('CRIM vs PRICE')
plt.xlabel('per capita crime rate by town')
plt.ylabel('Median value of owner-occupied homes in $1000s')

plt.show()
```

### Compare correlations of the features with each other and with PRICE

```python
# calculate the linear relationships between the variables
plt.figure(figsize = (15,6))
correlation_matrix = df.corr().round(2)
sns.heatmap(correlation_matrix, annot = True)
```

```python
# calculate the linear relationships between the variables
plt.figure(figsize = (15,6))
correlation_matrix = df.corr().round(2)

# Display only relevant portion of matrix
mask = np.zeros_like(correlation_matrix, dtype=bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(correlation_matrix, annot = True, mask=mask)
```

Independent variables (CRIM, INDUS, NOX, AGE, RAD, TAX, PTRATIO, LSTAT) are negatively correlated to the dependent variable (PRICE).  The rest of the independent variables are positively correlated to PRICE.

- RM has a strong positive correlation with PRICE (0.7)
- LSTAT has a high negative correlation with PRICE (-0.74).

When selecting features for a linear regression model it is important to check feature pairs for multicolinearity.
- RAD & TAX have a correlation of 0.91. They are strongly positively correlated to each other.
- DIS & AGE have a correlation of -0.75. They are strongly negatively correlated to each other.

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']
target = df['PRICE']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker='.')     # blue points
    plt.title("Variation in House prices")
    plt.xlabel(col)
    plt.ylabel('"House prices in $1000"')
```
